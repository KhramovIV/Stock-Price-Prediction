```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import pipeline
from transformers.pipelines.pt_utils import KeyDataset
from tqdm.auto import tqdm

device = torch.device('cuda') if torch.cuda.is_available() else torch.device("cpu")
print(f"device: {device}")
```

```python
df = pd.read_csv('news.csv')
texts = df['text'].to_list()

class TextDataset(Dataset):
    def __init__(self, texts):
        self.texts = texts
    def __len__(self):
        return len(self.texts)
    def __getitem__(self, idx):
        return self.texts[idx]

ds = TextDataset(texts)

model = pipeline(
    "sentiment-analysis",
    model="mxlcw/rubert-tiny2-russian-financial-sentiment", 
    device=device,
    batch_size=64
)

results = []
for out in tqdm(model(ds), total=len(ds)):
    results.append(out)

df['sentiment'] = [res['label'] for res in results]
df['sentiment_score'] = [res['score'] for res in results]
```

Сохраняем

```python
df.to_csv("../data/1-raw/sentiment.csv")
```
